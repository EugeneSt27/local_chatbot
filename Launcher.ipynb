# Запуск Локального AI Assistant
Этот блокнот позволяет запустить выбранный бэкенд (vLLM или Ollama) и UI прямо внутри вашей Jupyter среды.

```python
import subprocess
import time
import sys
import config

engine = config.ENGINE
model_path_or_name = config.MODEL_PATH_OR_NAME
port = config.PORT

print(f"Инициализация бэкенда '{engine}' на порте {port} для модели '{model_path_or_name}'...")

# Запускаем сервер в фоновом режиме через наш скрипт
backend_process = subprocess.Popen(
    [sys.executable, "start_backend.py"],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True
)

print("Ждем начала работы сервера (около 15 секунд)...")
time.sleep(15)
print("Бэкенд готов!")
```

```python
print("Запуск Streamlit UI...")
streamlit_process = subprocess.Popen(
    [sys.executable, "-m", "streamlit", "run", "app.py"]
)
print("UI должен открыться в новой вкладке браузера.")
```

### Остановка работы
Выполните нижнюю ячейку, когда закончите работу. Это закроет фоновые процессы.

```python
try:
    backend_process.terminate()
    streamlit_process.terminate()
    print("Все процессы успешно завершены.")
except:
    print("Процессы уже были завершены.")
```
